````python
# imports
from scraper import fetch_website_links,fetch_website_contents
from openai import OpenAI
import requests
from IPython.display import Markdown, display, update_display
````

````python
# constants
requests.get("http://localhost:11434").content
#MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA ="llama3.2:1b"
````

````python
# set up environment
OLLAMA_BASE_URL="http://localhost:11434/v1"
ollama=OpenAI(base_url=OLLAMA_BASE_URL,api_key='ollama')
````

````python
# here is the question; type over this to ask something new

question = """
Please explain what this  does and why:
why does we have feaver?
"""
````

````python
# Get gpt-4o-mini to answer, with streaming
user_promt="""your are a very helpful assistent.Your job is to give a details answer so in easy english world so that it can be understood by any one.Also your eager to help """
system_promt="""pls do your best to give ans to the question""" +question
````

````python
# Get Llama 3.2 to answer
messages=[
    {
        "role":"system","content":system_promt
    },
    {
        "role":"user","content":user_promt

    }

    
]
````

````python
response=ollama.chat.completions.create(model=MODEL_LLAMA,messages=messages)
````

````python
response.choices[0].message.content
````

````python
# Get gpt-4o-mini to answer, with streaming

stream = ollama.chat.completions.create(model=MODEL_LLAMA, messages=messages,stream=True)
    
response = ""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response += chunk.choices[0].delta.content or ''
    response = response.replace("```","").replace("markdown", "")
    update_display(Markdown(response), display_id=display_handle.display_id)
````
